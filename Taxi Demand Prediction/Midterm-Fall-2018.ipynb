Next, we would like to discover outliers. Do you see any?

```python
#Your Code here
stds = 1.0  # Number of standard deviation that defines 'outlier'.
z = taxi_data[:].groupby('vendor_id').transform(
    lambda group: (group - group.mean()).div(group.std()))
outliers = z.abs() > stds

df1 = taxi_data[outliers.any(axis=1)]
df1
```

## Finding Peak Hours

An interesting analysis could be find out the peak hours. This can be achieved if we count the number of trip for in each hour. You need to do the following:
- Plot the number of trips in each hour. 
- List the hours w.r.t number of trips in descending hour.

```python
import matplotlib.pyplot as plt
#Your Code here
hours = df1['pickup_datetime']
#hours = pd.DataFrame(hours)
hours = hours.dt.hour

unique_elements, counts_elements = np.unique(hours, return_counts=True)
plt.plot(counts_elements)

plt.title("Combined Statistics")
plt.xlabel("Hours")
plt.ylabel("Trips")

plt.show()

print -np.sort(-counts_elements)
```
